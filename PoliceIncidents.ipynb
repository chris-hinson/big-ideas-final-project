```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
crime = pd.read_csv("policeincidents.csv")
```

```python
crimesorted = crime.groupby('INCIDENTNEIGHBORHOOD').size().sort_values()
```

```python
crimesorted
```

```python
meancrime = crimesorted.mean()
print(meancrime)
```

```python
threshold = 0
neighborhoods = 0
for i in crimesorted:
    if i<meancrime:
        threshold = threshold+1
    neighborhoods = neighborhoods+1
    
print(threshold)
print(neighborhoods)
```

```python
crimesorted[61]
```

```python
tailnumber = neighborhoods - threshold
```

```python
worst = crimesorted.tail(tailnumber)
```

```python
worst.plot.bar(fontsize = 20, title = "Neighborhoods above average crime rate by number of crimes reported",figsize=(20,10))
plt.axhline(meancrime,color='k')
```
